# Tarefa
A tarefa proposta é executar diversos algoritmos de classificação e analizá-los de acordo com seus hiperparâmetros e qualidade. o código necessita de scikit-learn versão 1.0 no mínimo para funcionar, devido à função RocCurveDisplay.

## Imports

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn import preprocessing

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate

from sklearn.metrics import auc
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import accuracy_score

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.neighbors import KNeighborsClassifier

from sklearn.svm import SVC

from sklearn.naive_bayes import GaussianNB

from sklearn.preprocessing import StandardScaler

from sklearn.tree import DecisionTreeClassifier
```

## Pré-processamento de Dados

```python
df = pd.read_csv("koi_data.csv")
df.head()
```

```python
features = df.drop(['kepoi_name', 'koi_disposition'], axis=1)
labels_y = np.where(df['koi_disposition'] == 'CONFIRMED', 1, 0)
```

```python
# DATA NORMALIZATION
pre_processed_features = features.to_numpy()
scaler = StandardScaler().fit(pre_processed_features)
X = scaler.transform(pre_processed_features)
metrics = pd.DataFrame()
```

## Experimento Padrão

```python
def display_results(data, period):
    # DISPLAY PATTERN
    acc_str = "ACCURACY: " + ' '*10 + "{:.5f}\n".format(np.array(data).mean())
    std_str = "STANDARD DEVIATION: {:.5f}\n".format(np.array(data).std())
    buffer_str = '='*20 + '\n'
    display_str = buffer_str + period + ": \n" + acc_str + std_str + buffer_str
    
    print(display_str)
```

```python
def testbed(model, X, y, plot_title):
    true_positive_rates = []
    area_under_curves = []
    mean_fpr = np.linspace(0, 1, 100)
    
    scoring = ['accuracy', 'precision', 'recall', 'f1']
    results_training = [[],[],[],[],[]]
    results_testing = [[],[],[],[],[]]

    kf = KFold(n_splits=5)
    kf.get_n_splits(X)

    fig, ax = plt.subplots()

    # EPOCH
    for index, (train, test) in enumerate(kf.split(features)):
        model.fit(X[train], y[train])

        # PLOTTING EACH FOLD OF ROC CURVE
        roc_plot = RocCurveDisplay.from_estimator(
            model,
            X[test],
            y[test],
            name="Fold {}".format(index),
            ax=ax
        )

        # INTERPOLATE TRUE POSITIVE RATE
        interp_tpr = np.interp(mean_fpr, roc_plot.fpr, roc_plot.tpr)
        interp_tpr[0] = 0.0

        true_positive_rates.append(interp_tpr)
        area_under_curves.append(roc_plot.roc_auc)

        # PREDICT TRAINING
        y_pred = model.predict(X[train])
        accuracy = accuracy_score(y[train], y_pred)
        results_training[index].append(accuracy)

        # PREDICT TESTING
        y_pred = model.predict(X[test])
        accuracy = accuracy_score(y[test], y_pred)
        results_testing[index].append(accuracy)

    training_data = []
    testing_data  = []
    for index in range(5):
        training_data.append(np.array(results_training[index]).mean())
        testing_data.append(np.array(results_testing[index]).mean())
        
    # VISUAL RESPONSE AND GRAPHICS
    display_results(training_data, "TRAIN")
    display_results(testing_data, "TEST")

    mean_tpr = np.mean(true_positive_rates, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(area_under_curves)
    ax.plot(
        mean_fpr,
        mean_tpr,
        label="Mean ROC"
    )

    ax.set(
        xlim=[-0.05, 1.05],
        ylim=[-0.05, 1.05],
        title=plot_title
    )
    ax.legend(loc="lower right")
    plt.show()

    results = cross_validate(estimator=model, X=X, y=y, cv=5,scoring=scoring,return_train_score=True)
  
    metrics = {"Model Name": plot_title, "Results": results, "Mean False Positive Rate": mean_fpr, "Mean True Positive Rate": mean_tpr, "Mean Area Under Curve": mean_auc, "Standard AUC Deviation": std_auc}
    return metrics
```

## Algoritmos

### Naive Bayes
Esse classificador, como o nome já indica, é "ingênuo", pois assume que as features são independentes condicionalmente, e não é esperado que ele possua os melhores resultados dentre os classificadores testados. Por outro lado, é um modelo simples e rápido.

```python
model = GaussianNB()
NB_metrics = testbed(model, X, labels_y, 'Naive Bayes')
```

### Decision Tree
O algoritmo de Árvore de Decisão funciona com base em um esquema de galhos e folhas, onde as conclusões são obtidas ao final da árvore, ou seja, nas folhas. No caso, será testado valores de forma a observar o comportamento da árvore. Caso os resultados fossem insatisfatórios, eu executava incrementos maiores de profundidade, e caso o valor começasse a parecer aceitável, eu incrementava menos para observar a forma com que o algoritmo se comporta variando a profundidade.

```python
title = 'Decision Tree'
# For each testbed, the hiperparamenter is varied accordingly. The best overall result is selected to be saved

print("max_depth=3")
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
_ = testbed(model, X, labels_y, title)

print("max_depth=5")
model = DecisionTreeClassifier(criterion='entropy', max_depth=5)
DT_metrics = testbed(model, X, labels_y, title)

print("max_depth=6")
model = DecisionTreeClassifier(criterion='entropy', max_depth=6)
_ = testbed(model, X, labels_y, title)

print("max_depth=8")
model = DecisionTreeClassifier(criterion='entropy', max_depth=8)
_ = testbed(model, X, labels_y, title)
```

No caso, com profundidade 3 a acurácia e AUC são baixas, categorizando um possível caso de underfitting.

Foi testado com 2 níveis de profundidade extra, e é observado que com profundidade máxima de 5 o modelo possuem acurácia e AUC satisfatórios para um modelo em condições normais.

Então, adcionei mais 2 níveis de profundidade, e é observado que, apesar da acurácia estar melhor que anteriormente, a AUC é significativamente pior, podendo caracterizar um caso de overfitting. O mesmo acontece com 6 de profundidade máxima, indicando que 5 é o comprimento mais adequando para a árvore.

### SVM
O algoritmo SVM é utilizado para mapear o espaço do modelo entre classes, podendo ser um algoritmo bastante versátil para classificação, devido seu alto grau de customização. Isso se deve às diferentes funções de kernel utilizadas para alterar a forma como o algoritmo decide as classes. É testado 4 diferentes kernels para o problema proposto.

```python
# For each testbed, the hiperparamenter is varied accordingly. All results are saved for comparison

model = SVC(kernel="linear", probability=True)
LSVM_metrics = testbed(model, X, labels_y, "Linear SVM")

model = SVC(kernel="sigmoid", probability=True)
SSVM_metrics = testbed(model, X, labels_y, "Sigmoid SVM")

model = SVC(kernel="poly", probability=True)
PSVM_metrics = testbed(model, X, labels_y, "Poly SVM")

model = SVC(kernel="rbf", probability=True)
RSVM_metrics = testbed(model, X, labels_y, "RBF SVM")
```

Os kernels "sigmoid" e "polinomial" apresentam comportamentos insatisfatórios, visto que tanto a acurácia e a AUC não estão boas o suficiente para o problema. Ao mesmo tempo, os kernels "RBF" e "linear" possuem resultados bons, com acurácias e AUC similares, porém o kernel "linear" possui resultados levemente melhores.

### k-NN
O k-Nearest Neighbours classifica os dados a partir de um sistema de "votação", onde as entidades são atribuidas à classe mais comum aos k vizinhos do elemento. O principal hiperparâmetro desse método é o K, que é a quantidade de vizinhos próximos e é variado gradativamente nos testes.

```python
title = "k-NN"
# For each testbed, the hiperparamenter is varied accordingly. The best overall result is selected to be saved

print("k-NN with n_neighbors=1")
model = KNeighborsClassifier(n_neighbors=1)
_ = testbed(model, X, labels_y, title)
print("k-NN with n_neighbors=5")
model = KNeighborsClassifier(n_neighbors=5)
_ = testbed(model, X, labels_y, title)
print("k-NN with n_neighbors=10")
model = KNeighborsClassifier(n_neighbors=10)
knn_metrics = testbed(model, X, labels_y, title)
print("k-NN with n_neighbors=15")
model = KNeighborsClassifier(n_neighbors=15)
_ = testbed(model, X, labels_y, title)
print("k-NN with n_neighbors=20")
model = KNeighborsClassifier(n_neighbors=20)
_ = testbed(model, X, labels_y, title)
```

Com k = 1, é notável o comportamento de overfitting gerado, pois cada entidade teria sua própria classe, o que torna os testes horríveis. A partir desse ponto, o valor de K foi aumentando de 5 em 5, chegando ao pico por volta de 10, onde os valores de acurácia são os melhores, e com uma AUC média tão boa quanto a de k = 15.

### Random Forest
O algoritmo de Floresta Aleatória funciona a partir da classificação feita por diversas árvores, denominadas na implementação como "estimators", cujo a quantidade pode variar de acordo com o usuário. Sua classificação é dada por um processo de escolha da maioria, onde o valor escolhido é aquele que foi indicado pela maioria das árvores geradas pelo algoritmo.

```python
title = "Random Forest"
# For each testbed, the hiperparamenter is varied accordingly. The best overall result is selected to be saved

print("Random Forrest with n_estimators=3")
model = RandomForestClassifier(n_estimators=3)
_ = testbed(model, X, labels_y, title)
print("Random Forrest with n_estimators=5")
model = RandomForestClassifier(n_estimators=5)
_ = testbed(model, X, labels_y, title)
print("Random Forrest with n_estimators=10")
model = RandomForestClassifier(n_estimators=10)
RF_metrics = testbed(model, X, labels_y, title)
print("Random Forrest with n_estimators = 20")
model = RandomForestClassifier(n_estimators=20)
_ = testbed(model, X, labels_y, title)
print("Random Forrest with n_estimators=40")
model = RandomForestClassifier(n_estimators=40)
_ = testbed(model, X, labels_y, title)
```

O desempenho do método é muito bom, e é notável que com o aumento das árvores, melhor o método fica. Porém, a AUC diminui após 10 árvores, o que indica que essa pode ser a quantidade ideal de "estimators"

### Gradient Tree Boosting
O Gradient Tree Boosting, assim como o Random Forrest, é um algoritmo que utiliza árvores, mas nesse caso essas estruturas são aprendizes fracos, pois executam predicão fraca. Assim como o Random Forest, o hiperparâmetro variável são a quantidade de árvores do modelo, mas, diferentemente do classificador anterior, essas árvores são aprendizes fracos.

```python
title = "Gradient Boosting"
# For each testbed, the hiperparamenter is varied accordingly. The best overall result is selected to be saved

print("GTB with n_estimators=10")
model = GradientBoostingClassifier(n_estimators=10)
_ = testbed(model, X, labels_y, title)
print("GTB with n_estimators=20")
model = GradientBoostingClassifier(n_estimators=20)
_ = testbed(model, X, labels_y, title)
print("GTB with n_estimators=40")
model = GradientBoostingClassifier(n_estimators=40)
_ = testbed(model, X, labels_y, title)
print("GTB with n_estimators=80")
model = GradientBoostingClassifier(n_estimators=80)
_ = testbed(model, X, labels_y, title)
print("GTB with n_estimators=120")
model = GradientBoostingClassifier(n_estimators=120)
GTB_metrics = testbed(model, X, labels_y, title)
print("GTB with n_estimators=160")
model = GradientBoostingClassifier(n_estimators=160)
_ = testbed(model, X, labels_y, title)
```

Ao aumentar os valores, percebe-se que o desempenho começa a se estabilizar por volta de 80 árvores, mas lentamente melhora com o tempo. A partir de 80, o algoritmo já é ótimo, e sua AUC métida é a mesma, mas a acurácia de treino e teste continua a aumentar. Por isso, considero os valores ideais para o algoritmo entre 120 e 160 árvores.

# Comparações

```python
# Agregating all metrics together

metrics_df = pd.DataFrame(columns=["Model Name", "Results", "Mean False Positive Rate", "Mean True Positive Rate", "Mean Area Under Curve", "Standard AUC Deviation"])
metrics_df = metrics_df.append(NB_metrics, ignore_index=True)
metrics_df = metrics_df.append(DT_metrics, ignore_index=True)
metrics_df = metrics_df.append(LSVM_metrics, ignore_index=True)
metrics_df = metrics_df.append(SSVM_metrics, ignore_index=True)
metrics_df = metrics_df.append(PSVM_metrics, ignore_index=True)
metrics_df = metrics_df.append(RSVM_metrics, ignore_index=True)
metrics_df = metrics_df.append(knn_metrics, ignore_index=True)
metrics_df = metrics_df.append(RF_metrics, ignore_index=True)
metrics_df = metrics_df.append(GTB_metrics, ignore_index=True)
```

```python
fig, ax = plt.subplots()

trainpd = []
testpd = []
for index in range(9):
    mfpr = metrics_df["Mean False Positive Rate"][index]
    mtpr = metrics_df["Mean True Positive Rate"][index]
    ax.plot(
        mfpr,
        mtpr,
        label = metrics_df["Model Name"][index]
    )

    results =  metrics_df["Results"][index]
    train_precision_mean = results["train_precision"].mean() 
    test_precision_mean  = results["test_precision"].mean()
    train_precision_std = results["train_precision"].std() 
    test_precision_std  = results["test_precision"].std()

    trainpd.append((train_precision_mean, train_precision_std))
    testpd.append((test_precision_mean, test_precision_std))

metrics_df["Train Precision and Deviation"] = trainpd
metrics_df["Test Precision and Deviation"] = testpd

ax.set(
    xlim=[-0.05, 1.05],
    ylim=[-0.05, 1.05],
    title="Comparative Plot",
)
ax.legend(loc="lower right")
plt.show()
analyse_metrics = metrics_df.drop(metrics_df.columns[[1, 2, 3]], axis=1)
analyse_metrics
```

É possível observar claramente que quanto mais simples os algoritmos, pior eles performam na tarefa proposta. Os algoritmos com melhor AUC são o Random Forest, Gradient Boosting e o SVM com kernel linear e RBF. Isso indica que esses modelos tentem a classificar elementos de forma positiva, visto que quanto maior a AUC, melhor é a proporção de "True Positive Rate" para "False Positive Rate", que é representada pela curva ROC.

Em relação a presião, os melhores são o Random Forest e Gradien Boosting, tanto em treino quanto em teste. Coincidentemente, esses métodos são similares e são os mais complexos da seleção de modelos feita para esse trabalho prático, possuindo os maiores tempos de execução dos algoritmos presentes nessa análise.
